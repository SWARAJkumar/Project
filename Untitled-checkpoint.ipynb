```python
print(data['numerosity'].iloc[0]-data['numerosity'].iloc[1])
#print(data['numerosity'].iloc[:-1])
df=data['numerosity'].values[1:]-data['numerosity'].values[:-1]
print(df)
deletion_index=[]
dec_count=0
inc_count=0
for i in range(0,1998):
    if(df[i]>0):
        inc_count= inc_count+1
    elif (df[i]<0):
        dec_count= dec_count+1
        deletion_index.append(i+1)
print(inc_count, dec_count)
print(deletion_index)
```

```python
df=data['no_macro'].values[1:]-data['no_macro'].values[:-1]
print(df)
dec_count=0
inc_count=0
for i in range(0,1998):
    if(df[i]>0):
        inc_count= inc_count+1
    elif (df[i]<0):
        dec_count= dec_count+1
print(inc_count, dec_count)
```

```python
plt.plot(data['gameNo'],data['avg_fitness'])
for i in deletion_index:
    plt.axvline(x=i)
plt.title("Fitness")
plt.show()
```

Blue lines represent the game numbers where the deletion took place 

```python
plt.plot(data['gameNo'],data['avg_prediction_error'],color='r')
for i in deletion_index:
    plt.axvline(x=i)
plt.title("Prediction Error")
plt.show()
```
