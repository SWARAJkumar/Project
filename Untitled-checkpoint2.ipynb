```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
```

```python
df= pd.read_csv("L5_pop3k.csv")
df.head()
```

```python
ax=plt
plt.plot(df['gameNo'][10:],df['XCS'][10:])
plt.plot(df['gameNo'][10:],df['Minimax'][10:])
plt.plot(df['gameNo'][10:],df['Tie'][10:])
plt.legend()
plt.title("Level 5 Population size :3000 ")
plt.xlabel('Number of games', fontsize=18)
plt.ylabel('Win Percentage', fontsize=18)
plt.show()
```

The random spikes is due to deletion. When ever deletion happens--> spike

```python
df['Fitness'].iloc[:1000].hist(bins=100, stacked= True, label="0-1k")
df['Fitness'].iloc[1000:2000].hist(bins=100,stacked= True,label="1k-2k")
df['Fitness'].iloc[2000:3000].hist(bins=100,stacked= True,label="2k-3k")
df['Fitness'].iloc[3000:4000].hist(bins=100,stacked= True,label="3k-4k")
df['Fitness'].iloc[4000:5000].hist(bins=100,stacked= True,label="4k-5k")
df['Fitness'].iloc[5000:].hist(bins=100,stacked= True,label="5k-6k")
plt.legend()
plt.title("Histogram representing the fitness with train")
plt.show()
```

```python
df['result'].iloc[:1000].hist(bins=100, stacked= True, label="0-1k")
df['result'].iloc[1000:2000].hist(bins=100,stacked= True,label="1k-2k")
df['result'].iloc[2000:3000].hist(bins=100,stacked= True,label="2k-3k")
df['result'].iloc[3000:4000].hist(bins=100,stacked= True,label="3k-4k")
plt.legend()
plt.title("Histogram representing the fitness with train")
plt.show()
```

for 1k-2k transition fitness moves right which is positive but for the remaning training it is decreasing

```python
plt.plot(df['numerosity'],df['Prediction Error'])
plt.xlabel("numerosity")
plt.ylabel("prediction error")
plt.show()
```

When ever deletion takes place prediciton error shoots up and then decreases but there is an overall increase in the prediciton error

```python
colors = np.where(df.result=="black", 'b', 'r')
plt.scatter(df['Fitness'], df['Prediction Error'], s=10, c=colors)
plt.xlabel("fitness")
plt.show()
```

This shows neither the prediction error nor the fitness have any effect on winning as XCS wins across the spectrum

```python
data=df.copy()  
# data is a copy of original df, df shall not be used from this point onward for plotting purpose

df=data['numerosity'].values[1:]-data['numerosity'].values[:-1]
print(df)
deletion_index=[]
dec_count=0
inc_count=0
for i in range(0,5600):
    if(df[i]>0):
        inc_count= inc_count+1
        
    elif (df[i]<0):
        dec_count= dec_count+1
        deletion_index.append(i+1)
print(inc_count, dec_count)
print(deletion_index)
```

```python
df=data['Macro Classifier'].values[1:]-data['Macro Classifier'].values[:-1]
print(df)
dec_count=0
inc_count=0
for i in range(0,1998):
    if(df[i]>0):
        inc_count= inc_count+1
    elif (df[i]<0):
        dec_count= dec_count+1
print(inc_count, dec_count)
```

```python
plt.plot(data['gameNo'],data['Fitness'],color='b')
for i in deletion_index:
    plt.axvline(x=i)
plt.title("Fitness")
plt.xlabel("Games")
plt.ylabel("fitness")
plt.show()
```

```python
for i in deletion_index:
    plt.axvline(x=i, color='y')
plt.plot(data['gameNo'],data['Prediction Error'],color='k')
plt.title("Prediction Error")
plt.xlabel("Games")
plt.ylabel("error")
plt.show()
```
